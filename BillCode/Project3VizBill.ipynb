```python
import pandas as pd 
import requests
```

```python
pd.options.display.max_columns = 100
pd.options.display.max_rows = 999
```

```python
response = requests.get("https://www.ngdc.noaa.gov/hazel/hazard-service/api/v1/volcanoes?maxYear=2022&minYear=2000")
```

```python
the_json = response.json()
the_dicts = the_json["items"]

# print(the_dicts)
```

```python
df = pd.DataFrame(data=the_dicts)
```

```python
df
```

```python
df = pd.json_normalize(the_dicts)
```

```python
df
```

```python
df.columns
```

```python
df = df[["id", "name", "location", "country", "latitude", "longitude", "elevation", "year", "month", "day", "morphology", "vei", "deathsTotal", "missingTotal", "injuriesTotal", "damageMillionsDollarsTotal"]]
```

```python
df
```

```python
df.sort_values(by="year")
```

```python
df.describe()
```

```python
df["morphology"].value_counts().to_frame()
```

```python
df["country"].value_counts().to_frame()
```

```python
df.groupby("year")["id"].count().plot(figsize=(20,10), xticks=range(2000, 2023), ylabel="Number of Volcanoes", grid=True)
```

```python
df.groupby("year")["deathsTotal"].sum().plot(figsize=(20,10), xticks=range(2000, 2023), ylabel="Total Deaths", grid=True)
```

```python
df.groupby("year")["missingTotal"].sum().plot(figsize=(20,10), xticks=range(2000, 2023), ylabel="Total Missing", grid=True)
```

```python
df.groupby("year")["injuriesTotal"].sum().plot(figsize=(20,10), xticks=range(2000, 2023), ylabel="Total Injuries", grid=True)
```

```python
df.groupby("year")["damageMillionsDollarsTotal"].sum().plot(figsize=(20,10), xticks=range(2000, 2023), ylabel="Total Damage in $ (millions)", grid=True)
```

```python
df.groupby(["vei"])["id"].count().plot(figsize=(20,10), xticks=range(0, 9), ylabel="Number of Volcanoes", grid=True)
```

```python
df.groupby(["country"])["id"].count().plot(kind="bar", figsize=(20,10), ylabel="Number of Volcanoes", grid=True)
```
